```python
import warnings 
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='white',color_codes=True)
import pandas as pd
```

```python
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
df.info()
```

```python
df[df.columns[:8]].hist(figsize=(10,10))
```

```python
sns.distplot(df.Age)
```

```python
sns.FacetGrid(df, hue='Outcome',size=6).map(sns.kdeplot, "Age", shade=True, cut=True).add_legend()
sns.FacetGrid(df, hue='Outcome',size=6).map(sns.kdeplot, "Pregnancies", shade=True, cut=True).add_legend()
```

```python
corr = df[df.columns[:8]].corr()
print(corr.shape)
plt.figure(figsize=(14,14))
sns.heatmap(corr, cbar=True, square= True, fmt='.2f',annot=True,annot_kws={'size':15}, cmap='Greens')
```
